```python
import matplotlib.pyplot as plt

# Product names
products = ["Product A", "Product B", "Product C", "Product D", "Product E", "Product F"]

# Sales values
sales = [150, 200, 180, 220, 170, 210]

# Colors for each bar
colors = ["red", "blue", "green", "orange", "purple", "brown"]

# Create bar chart
plt.bar(products, sales, color=colors)

# Chart title
plt.title("Sales of 6 Products")

# X-axis label
plt.xlabel("Products")

# Y-axis label
plt.ylabel("Sales")

# Display chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Months of the year
months = ["Jan","Feb","Mar","Apr","May","Jun",
          "Jul","Aug","Sep","Oct","Nov","Dec"]

# Monthly sales data (example)
sales = [1200, 1500, 1700, 1600, 1800, 2000,
         2200, 2100, 1900, 2300, 2500, 2700]

# Create line chart
plt.plot(months, sales, marker='o', linewidth=2)

# Title
plt.title("Monthly Sales of a Company")

# Labels
plt.xlabel("Months")
plt.ylabel("Sales")

# Grid
plt.grid(True)

# Display chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Months
months = ["Jan", "Feb", "Mar", "Apr", "May", "Jun"]

# Sales data for three products
product_A = [120, 150, 170, 160, 180, 200]
product_B = [100, 130, 160, 150, 170, 190]
product_C = [90, 120, 140, 135, 150, 170]

# Plotting lines
plt.plot(months, product_A, marker='o', label="Product A")
plt.plot(months, product_B, marker='s', label="Product B")
plt.plot(months, product_C, marker='^', label="Product C")

# Chart details
plt.title("Sales of Three Products Over Six Months")
plt.xlabel("Months")
plt.ylabel("Sales")
plt.legend()
plt.grid(True)

# Display chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Departments
departments = ["CSE", "ECE", "Mechanical", "Civil"]

# Number of students
male = [40, 35, 30, 25]
female = [30, 25, 20, 15]

# Creating stacked bar chart
plt.bar(departments, male, label="Male")
plt.bar(departments, female, bottom=male, label="Female")

# Chart details
plt.title("Number of Male and Female Students in Departments")
plt.xlabel("Departments")
plt.ylabel("Number of Students")
plt.legend()

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Activities
activities = ["Study", "Sleep", "Exercise", "Entertainment"]

# Time spent in hours
time_spent = [6, 8, 2, 4]

# Explode effect (separate one slice)
explode = (0.1, 0, 0, 0)

# Creating exploded pie chart
plt.pie(time_spent, labels=activities, autopct='%1.1f%%', explode=explode)

# Title
plt.title("Daily Time Spent on Activities")

# Display chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Cities
cities = ["Delhi", "Mumbai", "Bangalore", "Chennai", "Kolkata"]

# Population (in millions)
population = [32, 20, 13, 11, 15]

# Area (in square km)
area = [1484, 603, 741, 426, 206]

# GDP values (used for bubble size)
gdp = [300, 250, 200, 180, 150]

# Create bubble chart
plt.scatter(area, population, s=[g*10 for g in gdp], alpha=0.6)

# Label each point with city name
for i in range(len(cities)):
    plt.text(area[i], population[i], cities[i])

# Chart details
plt.title("Population vs Area of Cities (Bubble Size = GDP)")
plt.xlabel("Area (sq km)")
plt.ylabel("Population (millions)")

# Display chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Years
years = [2019, 2020, 2021, 2022, 2023, 2024]

# Profit values (in millions)
profits = [10, 15, 20, 25, 30, 35]

# Create area chart
plt.fill_between(years, profits)

# Add title and labels
plt.title("Yearly Profit Growth of a Company")
plt.xlabel("Years")
plt.ylabel("Profit (in millions)")

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Time intervals
time = ["6AM", "9AM", "12PM", "3PM", "6PM", "9PM"]

# Electricity consumption (in units)
consumption = [2, 4, 6, 5, 7, 4]

# Create step chart
plt.step(time, consumption, where='mid')

# Add title and labels
plt.title("Electricity Consumption During Different Time Intervals")
plt.xlabel("Time of Day")
plt.ylabel("Electricity Consumption (Units)")

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Salary data (example)
dept_A = [35000, 40000, 42000, 45000, 47000]
dept_B = [30000, 32000, 34000, 36000, 38000]
dept_C = [45000, 48000, 50000, 52000, 55000]

# Combine data
data = [dept_A, dept_B, dept_C]

# Create box plot
plt.boxplot(data, labels=["Dept A", "Dept B", "Dept C"])

# Chart details
plt.title("Salary Distribution of Employees in Different Departments")
plt.xlabel("Departments")
plt.ylabel("Salary")

# Display the chart
plt.show()
```

```python
import matplotlib.pyplot as plt

# Exam scores of students in three classes
class_A = [65, 70, 75, 80, 85, 90]
class_B = [60, 62, 68, 72, 78, 82]
class_C = [70, 74, 78, 83, 88, 92]

# Combine data
data = [class_A, class_B, class_C]

# Create violin plot
plt.violinplot(data)

# Add labels
plt.xticks([1, 2, 3], ["Class A", "Class B", "Class C"])
plt.title("Distribution of Exam Scores in Different Classes")
plt.xlabel("Classes")
plt.ylabel("Exam Scores")

# Display the plot
plt.show()
```
